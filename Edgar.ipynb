```python
url = 'https://www.sec.gov/cgi-bin/browse-edgar?action=getcurrent&datea=&dateb=&company=&type=4&SIC=&State=&Country=&CIK=&owner=only&accno=&start=0&count=100'

with requests.Session() as se:
    se.headers = {
        "User-Agent": "Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/92.0.4515.131 Safari/537.36",
        "Accept-Encoding": "gzip, deflate",
        "Accept": "text/html,application/xhtml+xml,application/xml;q=0.9,image/webp,image/apng,*/*;q=0.8",
        "Accept-Language": "en"
    }

pages = 1
for page in range(1, pages+1):
    res = se.get(url.format(page),verify=False)
    soup = BeautifulSoup(res.text,"lxml")

k=soup.find_all('a')
link_list = []
filing_list = []

for i in k:
     i = str(i)
     if '-index.htm' in i:
        link = i.split("href=",1)[1]
        link = link.split('"')[1]
        link = 'https://www.sec.gov/' + link
        link_list.append(link)


        
link_length = len(link_list)      
for i in range(0,link_length):
    res = se.get(link_list[0].format(page),verify=False)
    soup = BeautifulSoup(res.text,"lxml")
    k=soup.find_all('a')
    link_length = []
    for i in k:
        i = str(i)
        link = i.split("a href=",1)[1]
        link = link.split('"')[1]
        if '/Archives/edgar/data' in i:
            link = 'https://www.sec.gov/' + link
            if 'xml' in str(link):
                link_length.append(link)
    x = min(link_length[0],link_length[1], key=len) 
    filing_list.append(x)
    del link_list[0]
    
unique_filings = []
for i in filing_list:
    if i not in unique_filings:
        unique_filings.append(i)
    
issuing_parties = []
reporting_periods = []
trading_symbols = []
securities = []
buy_or_sell = []
transactions = []
price_per_share = []
transaction_dates = []
all_footnotes = []
filing_length = len(unique_filings)
old_filings = []
for i in range(0,filing_length):
    res = se.get(unique_filings[0].format(page),verify=False)
    soup = BeautifulSoup(res.text,"lxml")
    issuer_name = str(soup.issuername)
    issuer_name = issuer_name.lstrip('<issuername>').rstrip('</issuername>')
    issuing_parties.append(issuer_name)
    period_of_report = str(soup.periodofreport)
    period_of_report = period_of_report.lstrip('<periodofreport>').rstrip('</periodofreport>')
    reporting_periods.append(period_of_report)
    trading_symbol = str(soup.issuertradingsymbol)
    trading_symbol = trading_symbol.lstrip('<issuertradingsymbol>').rstrip('</issuertradingsymbol>')
    trading_symbols.append(trading_symbol)
    security = str(soup.securitytitle)
    security = str(security.lstrip('<securitytitle>\n<value>').rstrip('a\n</securitytitle>'))
    security = str(security.rstrip('</value>\n<footnoteid id="F1"></footnoteid'))
    security = str(security.rstrip('</value>\n'))
    security = security.rstrip('</v')
    securities.append(security)
    shares_acquired_or_disposed = str(soup.transactionacquireddisposedcode)
    shares_acquired_or_disposed = str(shares_acquired_or_disposed.lstrip('<transactionacquireddisposedcode>').rstrip('/transactionacquireddisposedcode>'))
    shares_acquired_or_disposed = str(shares_acquired_or_disposed.rstrip('\n<footnoteid id="F1"></footnoteid>\n'))
    shares_acquired_or_disposed = shares_acquired_or_disposed.lstrip('\n<value>').rstrip('</value>\n<')
    buy_or_sell.append(shares_acquired_or_disposed)
    transaction_shares = str(soup.transactionshares)
    transaction_shares = str(transaction_shares.lstrip('<transactionshares>').rstrip('</transactionshares>'))
    transaction_shares = str(transaction_shares.lstrip('\n<value>').rstrip('</value>\n<footnoteid id="F1"></footnoteid>\n'))
    transaction_shares = transaction_shares.rstrip('</value>\n<footnoteid id="F2')
    transactions.append(transaction_shares)
    transaction_price_per_share = str(soup.transactionpricepershare)
    transaction_price_per_share = str(transaction_price_per_share.lstrip('<transactionpricepershare>').rstrip('</transactionpricepershare>'))
    transaction_price_per_share = str(transaction_price_per_share.lstrip('\n<footnoteid id="F2"').rstrip('\n</footnoteid>'))
    transaction_price_per_share = str(transaction_price_per_share.lstrip('\n<footnoteid id="F1"')).rstrip('\n<footnoteid id="F2"')
    transaction_price_per_share = transaction_price_per_share.lstrip('\n<value>').rstrip('</value>\n')
    price_per_share.append(transaction_price_per_share)
    transaction_date = str(soup.transactiondate)
    transaction_date = str(transaction_date.lstrip('<transactiondate>').rstrip('</transactiondate>'))
    transaction_date = transaction_date.lstrip('\n<value>').rstrip('</value>\n')
    transaction_dates.append(transaction_date)
    footnotes = str(soup.footnotes)
    footnotes = str(footnotes.lstrip('<footnotes>').rstrip('</footnotes>'))
    footnotes = footnotes.lstrip('\n<footnote id="F1">').rstrip('</footnote>\n')
    all_footnotes.append(footnotes)
    old_filings.append(unique_filings[0])
    del unique_filings[0]
```
