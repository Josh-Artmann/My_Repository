## Grab all links from homepage

```python
url = 'https://www.wsj.com/'
with requests.Session() as se:
    se.headers = {
        "User-Agent": "Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/92.0.4515.131 Safari/537.36",
        "Accept-Encoding": "gzip, deflate",
        "Accept": "text/html,application/xhtml+xml,application/xml;q=0.9,image/webp,image/apng,*/*;q=0.8",
        "Accept-Language": "en"
    }
pages = 1
for page in range(1, pages+1):
    res = se.get(url.format(page),verify=False)
    soup = BeautifulSoup(res.text,"lxml")

k=soup.find_all('a')
link_list = []
for i in k:
    i = str(i)
    if 'WSJTheme--headlineText' in i:
        link = i.split("href=",1)[1]
        link = link.split('"')[1]
        print(link)
        link_list.append(link)
        time.sleep(0)
```

## for loop to get article information from links 

```python
list_2 = ['wsj']
link_list = [x for x in link_list if
              all(y in x for y in list_2)]
titles = []
sub_heads = []
datetimes = []
for url in link_list:
    with requests.Session() as se:
        se.headers = {
            "User-Agent": "Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/92.0.4515.131 Safari/537.36",
            "Accept-Encoding": "gzip, deflate",
            "Accept": "text/html,application/xhtml+xml,application/xml;q=0.9,image/webp,image/apng,*/*;q=0.8",
            "Accept-Language": "en"
    }
        res_1 = se.get(url.format(page),verify=False)
        soup_1 = BeautifulSoup(res_1.text, 'html.parser')
        title = soup_1.find_all('title')
        title = str(title)
        title = title.rstrip('- WSJ</title>]').lstrip('[<title>')
        titles.append(title)
        sub_head = soup_1.find("h2",{"class":"sub-head"})
        sub_head = str(sub_head)
        sub_head = sub_head.rstrip('</h2>').lstrip('<h2 class="sub-head" itemprop="description">')
        sub_heads.append(sub_head)
        datetime = soup_1.find("time",{"class":"timestamp article__timestamp flexbox__flex--1"})
        datetime = str(datetime)
        datetime = datetime.rstrip('\n      </time>').lstrip('<time class="timestamp article__timestamp flexbox__flex--1">\n        ')
        datetimes.append(datetime)
        time.sleep(0)
```

## Create Dataframe (lfg)

```python
data = {'link': link_list, 'title': titles, 'sub_head': sub_heads, 'datetime': datetime}
wsj = pd.DataFrame(data)
wsj
```

## Searching google 

```python
import requests
import bs4
text= 'lithium+americas'
url = 'https://www.google.com/search?q='+text+'&safe=active&tbm=nws'
request_result=requests.get( url,verify=False )
soup = bs4.BeautifulSoup(request_result.text,
                         "html.parser")
```

```python
heading_object=soup.find_all( 'h3' )
goog_titles = []
for info in heading_object:
    title = info.getText()
    goog_titles.append(title)
```

```python
goog_links = soup.find_all('a')
filtered_goog_links = []
for link in goog_links:
    link = str(link)
    if link.startswith('<a href="/url?'):
        filtered_goog_links.append(link)
filtered_goog_links = filtered_goog_links[:-2]
filtered_goog_links = [i for a, i in enumerate(filtered_goog_links) if  a%2 == 0]
```

```python
last_goog_links = []
for link in filtered_goog_links:
    link = link.split('"><h3 class')[0]
    link = link.split('url?q=')[1]
    last_goog_links.append(link)
```

```python
data_goog = {'link': last_goog_links, 'title': goog_titles}
goog = pd.DataFrame(data_goog)
goog
```
